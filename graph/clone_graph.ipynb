```python
class Solution:
    def cloneGraph(self, node: Optional['Node']) -> Optional['Node']:

        visited = {}





        def dfs(node):
            if not node :
               return None
        
            if node in visited:
               return visited[node]
        

            clone = Node(node.val)

            visited[node] = clone
            for neighbor in node.neighbors:
            
                clone.neighbors.append(dfs(neighbor))
            
            return clone






           

        

        return dfs(node)
```
