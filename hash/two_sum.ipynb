```python
def twoSum(nums,target):

    seen = {}

    for i, num in enumerate(nums):

        complete = target - num



        if complete in seen:
            
            return [i, seen[complete]]
        

        seen[num] = i


```

```python
nums = [2, 7, 11, 15]
target = 9
print(twoSum(nums, target))
```
