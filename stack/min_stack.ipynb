```python
class MinStack:

    def __init__(self):
       
        self.stack = []
        self.min_stack = []

    def push(self, val):
       

        self.stack.append(val)

        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val) 
        

    def pop(self):
        
        
        rid = self.stack.pop()

        if rid == self.min_stack[-1]:
            self.min_stack.pop()

        return rid 
        

    def top(self):
        
        top = self.stack[-1]

        return top
        

    def getMin(self):
       
        return self.min_stack[-1]
```

```python
s = MinStack()
s.push(3)
s.push(1)
s.push(5)
s.push(0)
print(s.getMin())  # 1
s.pop()
print(s.getMin())  # 1 ✅
s.pop()
print(s.getMin())  # 3 ✅
```
